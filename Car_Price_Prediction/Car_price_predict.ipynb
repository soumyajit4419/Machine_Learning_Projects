## Applying ML Algorithms 

```python
app=[df2,fuelsystem,enginelocation,drivewheel,aspiration,fueltype,carcompany]
```

```python
df2=pd.concat(app,axis=1)
```

```python
df2.head()
```

```python
x=df2.drop('price',axis=1)
x
```

```python
y=df2['price']
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.svm import SVR
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import ShuffleSplit
```

```python
cv=ShuffleSplit(n_splits=10,test_size=0.3,random_state=42)
```

```python
print(cross_val_score(LinearRegression(),x,y,cv=cv))
```

```python
print(cross_val_score(LinearRegression(),x,y,cv=cv).mean())
```

```python
print(cross_val_score(SVR(kernel='linear',degree=3),x,y,cv=cv))
```

```python
print(cross_val_score(SVR(kernel='linear',degree=3),x,y,cv=cv).mean())
```

```python
print(cross_val_score(DecisionTreeRegressor(),x,y,cv=cv))
```

```python
print(cross_val_score(DecisionTreeRegressor(),x,y,cv=cv).mean())
```

```python
print(cross_val_score(RandomForestRegressor(),x,y,cv=cv))
```

```python
print(cross_val_score(RandomForestRegressor(),x,y,cv=cv).mean())
```

#### Random Forest Regressor gives the max accuracy so making its model and testing its error

```python
from sklearn.model_selection import train_test_split
```

```python
model=RandomForestRegressor()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
r2_score(y_test,prediction)
```
