## Applying Clustering Techniques 

### Hierarchical Clustering

```python
hier_cluster=linkage(x_standard,method='ward')
```

```python
plt.figure(figsize = (18,9))
dendrogram(hier_cluster)
plt.show()
```

### KMeans Clustering

```python
sse=[]
for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit_predict(x_standard)
    a=km.inertia_
    sse.append(a)
```

```python
p=range(1,10)
plt.figure(figsize=(16,7))
plt.plot(p,sse,marker = 'o', linestyle = '--')
plt.show()
```

### It can be seen from the graph that min cluster is 4. 

```python
km=KMeans(n_clusters=4)
prediction=km.fit_predict(x_standard)
```

```python
km_normal=df.copy()
```

```python
km_normal['cluster']=prediction
```

```python
km_normal.head()
```

```python
df1=km_normal[km_normal['cluster']==0]
df2=km_normal[km_normal['cluster']==1]
df3=km_normal[km_normal['cluster']==2]
df4=km_normal[km_normal['cluster']==3]
```

```python
plt.figure(figsize=(14,9))
sns.scatterplot(x='Age',y='Income',data=df1,color='blue')
sns.scatterplot(x='Age',y='Income',data=df2,color='green')
sns.scatterplot(x='Age',y='Income',data=df3,color='red')
sns.scatterplot(x='Age',y='Income',data=df4,color='purple')
plt.show()
```

Mostly the segments are overlapping

```python
purchase_df.head()
```

### Bulding Logistic Regression Model for Purchase Probability

```python
purchase_df['Mean_price']=(purchase_df['Price_1']+purchase_df['Price_2']+purchase_df['Price_3']+purchase_df['Price_4']+purchase_df['Price_5'])/5
```

```python
x=purchase_df['Mean_price']
x=pd.DataFrame(x)
x
```

```python
y=purchase_df['Incidence']
y
```

```python
pr_model=LogisticRegression()
```

```python
pr_model.fit(x,y)
```

```python
pr_model.coef_
```

### Calculating Purchase Probability

```python
a=purchase_df[['Price_1','Price_2','Price_3','Price_4','Price_5','Mean_price']]
a.describe()
```

 We can see that the values ragnges between 1.1 to 2.8

```python
price_range=np.arange(0.5,3.5,0.01)
price_range
```

```python
price_range_df = pd.DataFrame(price_range)
```

```python
price_range_df
```

```python
prediction=pr_model.predict_proba(price_range_df)
#It gives two value the probability of no purchase and probability of purchase
```

```python
purchase_probability=prediction[:,1]
purchase_probability
```

```python
price_range_df['Purchase_probability']=purchase_probability
```

```python
price_range_df=price_range_df.rename(columns={0:"Price"})
price_range_df.head()
```

### Calculating Price Elasticity

```python
coef=pr_model.coef_[0][0]
coef
```

```python
pe=coef*price_range_df['Price']*(1-price_range_df['Purchase_probability'])
pe
```

```python
price_range_df['Price_elasticity']=pe
```

```python
price_range_df.head()
```

```python
plt.figure(figsize = (9, 6))
plt.plot(price_range_df['Price'],price_range_df['Price_elasticity'] ,color='red')
plt.xlabel('Price')
plt.ylabel('Elasticity')
plt.title('Price Elasticity of Purchase Probability')
```

### Building Multinomial Logistic Regression Model for all brands

```python
brand_model.fit(x,y)
```

```python
brand_model.coef_
```

```python
coef=pd.DataFrame(brand_model.coef_)
coef
```

```python
coef=coef.rename(columns={0:"Price1",1:"Price2",2:"Price3",3:"Price4",4:"Price5"})
coef['Brand']=["Brand1","Brand2","Brand3","Brand4","Brand5"]
coef=coef.set_index('Brand')
coef
```

### As brand 5 is the most expensive brand we predict the elasticty for its own price

### <center>Own Brand Price probability and elasticty of brand 5</center>

### Own Brand 5 Probability

```python
price_range=np.arange(0.5,3.5,0.01)
```

```python
brand_5_prediction=pd.DataFrame(index=np.arange(price_range.size))
brand_5_prediction['Price_1']=brand['Price_1'].mean()
brand_5_prediction['Price_2']=brand['Price_2'].mean()
brand_5_prediction['Price_3']=brand['Price_3'].mean()
brand_5_prediction['Price_4']=brand['Price_4'].mean()
brand_5_prediction['Price_5']=price_range
brand_5_prediction
```

```python
prediction=brand_model.predict_proba(brand_5_prediction)
prediction
```

```python
brand_predict_df=prediction[:,4]
brand_predict_df=pd.DataFrame(brand_predict_df)
brand_predict_df=brand_predict_df.rename(columns={0:"Own brand 5 probability"})
brand_predict_df
```

### Own Brand 5 Elasticity

```python
purchase_quantity_df_s0=purchase_df[purchase_df['Incidence']==1]
purchase_quantity_df_s0=purchase_quantity_df_s0[purchase_quantity_df_s0['segment']==0]
purchase_quantity_df_s0=pd.get_dummies(purchase_quantity_df_s0,columns=['Brand'],prefix='Brand')
purchase_quantity_df_s0['Price_Incidence'] = (purchase_quantity_df_s0['Brand_1'] * purchase_quantity_df_s0['Price_1'] +
                                           purchase_quantity_df_s0['Brand_2'] * purchase_quantity_df_s0['Price_2'] +
                                           purchase_quantity_df_s0['Brand_3'] * purchase_quantity_df_s0['Price_3'] +
                                           purchase_quantity_df_s0['Brand_4'] * purchase_quantity_df_s0['Price_4'] +
                                           purchase_quantity_df_s0['Brand_5'] * purchase_quantity_df_s0['Price_5'] )
x = purchase_quantity_df_s0[['Price_Incidence']]
y = purchase_quantity_df_s0['Quantity']
model_quantity_s0= LinearRegression()
model_quantity_s0.fit(x,y)
model_coef=model_quantity_s0.coef_
coef=model_coef[0]
```
